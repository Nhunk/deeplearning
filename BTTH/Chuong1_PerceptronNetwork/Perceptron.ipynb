## 1. Perceptron nguyn thu(Binary classifier)

1 Perceptron c th gii quyt c cc vn  n gin nh classification,..


c gii thiu bi Frank Rosenblatt nm 1957



### Xy dng 1 perceptron, 2 ng vo, 1 bias, 1 output vi cc trng s ngu nhin
- Kim tra du ca ng ra trn R2
- Thay i trng s v kim tra s thay i ca du trn R2


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

w = np.random.random((3,1)) -.5
def perceptron(x,w):
    xx=np.vstack((np.array([[1]]),x))
    y=w.T @ xx  # @ l tch v hng(dot produce)
    if y>0:
        return 1
    else:
        return 0

for x1 in np.arange(-5,5,.2):
    for x2 in np.arange(-5,5,.2):
        
        x = np.array([[x1,x2]]) # Thm ngoc  chuyn v 
        x=x.T #Chuyn v ma trn X

        a = perceptron(x,w)
        if a==1:
            plt.plot(x1,x2,'bo')
        else:
            plt.plot(x1,x2,'ro')
```

```python
# D liu hun luyn
X_train = np.array([[-1,2],
                    [1, 0], 
                    [0, 1],
                    [2, 0] ,
                    [1, 1], 
                    [3, 2]])  # Cc im d liu
y_train = np.array([0, 0, 0, 1, 1, 1])  # Nhn mong mun

# Thm bias vo X_train
X_train_bias = np.hstack((np.ones((X_train.shape[0], 1)), X_train))  # Thm ct 1 cho bias

# Hm hun luyn Perceptron
def train_perceptron(X_train, y_train, w, learning_rate =0.1, epochs=100):
    for epoch in range(epochs):
        for i in range(X_train.shape[0]):
            x = X_train[i].reshape(-1, 1)  # Chuyn i x thnh ct vector
            y_pred = perceptron(x[1:], w)  # B bias khi truyn vo hm perceptron
            if y_pred != y_train[i]:  # Cp nht trng s khi phn loi sai
                w = w + learning_rate*(y_train[i] - y_pred) * x
    return w

# Hun luyn Perceptron
w = train_perceptron(X_train_bias, y_train, w)

# V  th phn loi
for x1 in np.arange(-5, 5, 0.2):
    for x2 in np.arange(-5, 5, 0.2):
        x = np.array([[x1, x2]])  # To im d liu
        x = x.T  # Chuyn v ma trn x
        a = perceptron(x, w)
        if a == 1:
            plt.plot(x1, x2, 'bo', markersize=2.5)  # im lp 1
        else:
            plt.plot(x1, x2, 'ro', markersize=2.5)  # im lp 0

# V cc im hun luyn
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='coolwarm', marker='o', edgecolor='k', s=100)
plt.title('Perceptron Classification')
plt.show()
```

## 2. Perceptron ci tin (multi-label classifier)

```python
def perceptron_multi(X,W):
    X=np.array([X])
    X=X.T
    xx = np.vstack((X, np.array([[1]]))) # thm s 1 vo ct cui hay cn gi l bias  ng tuyn tnh xa ri gc to 
    y=W @ xx  # @ l tch v hng(dot produce)
    return y

X = np.array([[-1, -1], 
              [-2, 0], 
              [0, 1], 
              [2, 1] ]) 
t = np.array([0, 1, 2, 2])
W = np.array([[0,-1,0], 
              [1, 1, 0], 
              [0, 1,0]]) 

color = 'rgb'
plt.figure()

for i in range(X.shape[0]):
    plt.plot(X[i,0], X[i,1], color[t[i]]+'o' )
```

```python
epoch = 3
count = 0
for _ in range(epoch):
    for i in np.arange(X.shape[0]):
        out = perceptron_multi(X[i],W) # Tnh ton u ra
        xx = np.vstack((np.array([X[i]]).T, np.array([[1]]))) # thm s 1 vo ct cui
        flag = 'D'

        if np.argmax(out) != t[i]:
            flag = 'S'
            W[t[i]] = W[t[i]]+ xx.T

            for k in range(W.shape[0]):
                if out[k] == max(out):
                    W[k] =W[k]- xx.T

        print(W[0], W[1], W[2], '[X', i+1, 'L', t[i]+1,']', xx.T[0], out.T, flag)

        if flag =='D': count = count +1
        else: count = 0
        if count == 4: break
    if count == 4: break
```

```python
# V  TH PHN LOI 2D

plt.figure()
for x1 in np.arange(-5, 5, 0.2):
    for x2 in np.arange(-5, 5, 0.2):
        x = np.array([x1, x2])
        out = perceptron_multi(x, W)  # Tnh ton u ra
        label = np.argmax(out)  # Nhn d on
        plt.plot(x1,x2,color[label]+'.',markersize=3)  # V im d on

for i in range(X.shape[0]):
    plt.plot(X[i,0], X[i,1], color[t[i]]+'.',markersize = 15) # V cc im hun luyn
```

```python
# V  th phn loi 3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40, -60)  # Thit lp gc nhn 3D

for x1 in np.arange(-5, 5, 0.2):
    for x2 in np.arange(-5, 5, 0.2):
        x = np.array([x1, x2])
        out = perceptron_multi(x, W)  # Tnh ton u ra
        label = np.argmax(out)  # Nhn d on
        plt.plot(x1,x2,out[0],'r.')
        plt.plot(x1,x2,out[1],'g.')
        plt.plot(x1,x2,out[2],'b.')
plt.show()
```

### 3. Perceptron v cc vn  phi tuyn



- <b>D liu khng tc bng ng tuyn tnh</b>: Thay v dng mt phng(tuyn tnh)  phn bit cc vng d liu th khi xut hin nhng im ngoi lai kh c th kim sot, thay vo  c th s dng mt cong  linh hot hn trong vic phn loi


- <b>D liu phn tch bng ng bin phi tuyn c khng: </b> Thm cc kernel nh x1^2, x1*x2,... vo input ban u x1, x2  tnh sigma (cn gi l linear combinations) => <i>Mc ch l to ra b trng s w sao cho khp vi b d liu</i>

```python
# Hm perceptron phi tuyn
def perceptron_phituyen(x, w):
    # Thm cc kernel phi tuyn vo input
    xx = np.array([x[0], x[1], x[0]**2, x[1]**2, x[0]*x[1], 1])  # Thm bias
    y = w @ xx  # Tch v hng
    return y

# D liu hun luyn
X = np.array([[-1, -1], 
              [-2, 0], 
              [0, 1], 
              [2, 1]])
t = np.array([0, 1, 2, 2])  # Nhn
W = np.random.random((3, 6)) - 0.5  # Trng s ngu nhin vi kch thc ph hp (3 lp, 5 kernel + bias)

color = 'rgb'

# Hun luyn perceptron phi tuyn
epoch = 3
count = 0
for _ in range(epoch):
    for i in range(X.shape[0]):
        out = perceptron_phituyen(X[i], W)  # Tnh ton u ra
        xx = np.array([X[i, 0], X[i, 1], X[i, 0]**2, X[i, 1]**2, X[i, 0]*X[i, 1], 1])  # Thm kernel v bias
        flag = 'D'

        if np.argmax(out) != t[i]:  # Nu d on sai
            flag = 'S'
            W[t[i]] = W[t[i]] + xx  # Cp nht trng s cho lp ng
            for k in range(W.shape[0]):
                if k != t[i] and out[k] == max(out):  # Gim trng s cho lp sai
                    W[k] = W[k] - xx

        print(W[0], W[1], W[2], '[X', i+1, 'L', t[i]+1, ']', xx, out, flag)

        if flag == 'D':
            count += 1
        else:
            count = 0
        if count == 4:
            break
    if count == 4:
        break

# V  th phn loi 2D
plt.figure()
for x1 in np.arange(-5, 5, 0.2):
    for x2 in np.arange(-5, 5, 0.2):
        x = np.array([x1, x2])
        out = perceptron_phituyen(x, W)  # Tnh ton u ra
        label = np.argmax(out)  # Nhn d on
        plt.plot(x1, x2, color[label] + '.', markersize=5)  # V im d on

# V cc im hun luyn
for i in range(X.shape[0]):
    plt.plot(X[i, 0], X[i, 1], color[t[i]] + 'o', markersize=10)  # V cc im hun luyn

plt.title('Perceptron Phi Tuyn')
plt.show()
```
