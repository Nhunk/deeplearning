-	Ma trận BoW thông thường (đếm số lần xuất hiện của từ trong "vocab")
-	Ma trận TF
-	Ma trận TF-IDF
-	Tìm một số bài viết mới, phân loại xem thuộc chủ đề nào theo phương pháp Naive Bayes.


```python
data = [
  {
    "comment": "This is a fantastic product, I really love it!",
    "sentiment": "like"
  },
  {
    "comment": "Interesting. I'm not sure what to think yet.",
    "sentiment": "neutral"
  },
  {
    "comment": "Absolutely terrible, a complete waste of money.",
    "sentiment": "dislike"
  },
  {
    "comment": "It's okay, nothing particularly special though.",
    "sentiment": "neutral"
  },
  {
    "comment": "Wow! This exceeded my expectations by far.",
    "sentiment": "like"
  },
  {
    "comment": "I don't really have strong feelings about this.",
    "sentiment": "neutral"
  },
  {
    "comment": "This is not good. Very disappointed with it.",
    "sentiment": "dislike"
  },
  {
    "comment": "Pretty good overall, I would recommend it.",
    "sentiment": "like"
  }
]
```

```python
full_docx=""
for item in data:
    full_docx = full_docx + " " + item.get("comment")
    print(item.get("comment"),'_________',item.get("sentiment"))    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
s_data = nlp(full_docx)
clean_tokens = [token.text for token in s_data if not token.is_stop 
                                                    and not token.is_punct 
                                                    and not token.is_space
                                                    and token.has_vector]

# Kết quả
print("Cleaned tokens:")
print(clean_tokens)
corpus = [item["comment"] for item in data]
labels = [item["sentiment"] for item in data]

# Biểu diễn BoW
vectorizer_bow = CountVectorizer()
X_bow = vectorizer_bow.fit_transform(corpus)

# In từ điển
print("Vocabulary:", vectorizer_bow.get_feature_names_out())
print("BoW Matrix:\n", X_bow.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(corpus)

print("TF-IDF Matrix:\n", X_tfidf.toarray())
print("Vocabulary:", vectorizer_tfidf.get_feature_names_out())
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# Tạo mô hình với Tfidf + Naive Bayes
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(corpus, labels)

# Dự đoán bài viết mới
new_comments = [
    "I think this product is amazing!",
    "Not impressed. Very bad experience.",
    "It's alright, could be better."
]

predictions = model.predict(new_comments)
for text, label in zip(new_comments, predictions):
    print(f"'{text}' → Predicted: {label}")
```

```python
new_test_comments = [
    "I hate this product, it's the worst I've ever used.",       # -> dislike
    "I like how simple and effective it is.",                    # -> like
    "This product didn't meet my expectations at all.",         # -> dislike
    "Feels okay. Not too bad, not too good.",                   # -> neutral
    "Absolutely love it! Totally recommend.",                   # -> like
    "This is just average, nothing surprising.",                # -> neutral
    "Terrible design and awful performance.",                   # -> dislike
    "I'm so impressed, it's fantastic!",                        # -> like
    "Meh. It's not great, but it's not horrible either.",       # -> neutral
    "Useful and works exactly as expected.",              # -> like
    "Doesn't work as advertised. Really bad.",                  # -> dislike
]
predictions = model.predict(new_test_comments)
for text, label in zip(new_test_comments, predictions):
    print(f"'{text}' → Predicted: {label}")
```
