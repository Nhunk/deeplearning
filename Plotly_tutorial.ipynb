```python
import numpy as np
import plotly.graph_objects as go

# To d liu trc x, y
x = np.linspace(-5, 5, 50)  # T -5 n 5, chia thnh 50 im
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)    # To li ta 

# Hm z = x^2 + y^2
z = x**2 + y**2

# To biu  3D
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])

# Hin th biu  trong trnh duyt
fig.show()
```
