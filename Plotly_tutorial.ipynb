```python
import numpy as np
import plotly.graph_objects as go

# Tạo dữ liệu trục x, y
x = np.linspace(-5, 5, 50)  # Từ -5 đến 5, chia thành 50 điểm
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)    # Tạo lưới tọa độ

# Hàm z = x^2 + y^2
z = x**2 + y**2

# Tạo biểu đồ 3D
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])

# Hiển thị biểu đồ trong trình duyệt
fig.show()
```
